## ¿Cuáles son los docentes que tienen el mayor Salario y para qué carrera trabaja?
Para la siguiente operación se quiere conocer El Nombre, La Carrera, el centro y El salario de 7 docentes con el mayor salario

```python
th_borrado_Docente = th.merge(df2_Docente, how="inner", on="Id_Docente").drop(["Id_Nota", "Id_Estudiante"], axis = 1)
th_salario_Docente = th_borrado_Docente.sort_values(by=["Salario"], ascending=False).head(10).iloc[:,[2,3,4,5,6]]
th_salario_Docente
```
